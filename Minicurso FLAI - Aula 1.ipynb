```python
fifa = selecoes['PontosRankingFIFA'] #transformação linear da escala numérica
a, b = min(fifa), max(fifa)
fa, fb = 0.15, 1
b1 = (fb - fa)/(b - a)
b0 = fb - b*b1
forca = b0 + b1*fifa
```

```python
def MediasPoisson(selecao1, selecao2): #Média distribuição de Poisson
    forca1 = forca[selecao1]
    forca2 = forca[selecao2]
    mgols = 2.75
    l1 = mgols*forca1/(forca1 + forca2)
    l2 = mgols - l1

    return l1, l2
```

```python
def Resultado(gols1, gols2):
    if gols1 > gols2:
        resultado = 'V'
    elif gols1 < gols2:
        resultado = 'D'
    else:
        resultado = 'E'
    
    return resultado
```

```python
def Pontos(gols1, gols2):
    resultado = Resultado(gols1, gols2)
    if resultado == 'V':
        pontos1, pontos2 = 3, 0
    elif resultado == 'D':
        pontos1, pontos2 = 0, 3
    else:
        pontos1, pontos2 = 1, 1

    return [pontos1, pontos2, resultado]
```

```python
def Jogo(selecao1, selecao2):
    l1, l2 = MediasPoisson(selecao1, selecao2)
    gols1 = int(np.random.poisson(lam = 1, size = 1))
    gols2 = int(np.random.poisson(lam = 1, size = 1))
    saldo1 = gols1 - gols2
    saldo2 = -saldo1
    pontos1, pontos2, resultado = Pontos(gols1, gols2)
    placar = '{}x{}'.format(gols1, gols2)
    
    return [gols1, gols2, saldo1, saldo2, pontos1, pontos2, resultado, placar]
```

```python
def Distribuicao(media):
    probs = []
    for i in range(7):
        probs.append(poisson.pmf(i, media))
    probs.append(1 - sum(probs))
    return pd.Series(probs, index = ['0', '1', '2', '3', '4', '5', '6', '7+'])
```

```python
def ProbabilidadesPartida(selecao1, selecao2):
    l1, l2 = MediasPoisson(selecao1, selecao2)
    d1, d2 = Distribuicao(l1), Distribuicao(l2)
    matriz = np.outer(d1, d2)
    vitoria = np.tril(matriz).sum() - np.trace(matriz) #Soma triângulo inferior
    derrota = np.triu(matriz).sum() - np.trace(matriz) #Soma triângulo superior
    empate = 1 - (vitoria + derrota)

    probs = np.around([vitoria, empate, derrota], 3)
    probsp = [f'{100*i:.1f}%' for i in probs]

    nomes = ['0', '1', '2', '3', '4', '5', '6', '7+']
    mostrarMatriz = pd.DataFrame(matriz, columns = nomes, index = nomes)
    mostrarMatriz.index = pd.MultiIndex.from_product([[selecao1], mostrarMatriz.index])
    mostrarMatriz.columns = pd.MultiIndex.from_product([[selecao2], mostrarMatriz.columns])

    output = {'Seleção 1':selecao1, 'Seleção 2':selecao2, 'Força 1':forca[selecao1],
            'Força 2':forca[selecao2], 'Média 1':l1, 'Média 2':l2, 'Probabilidades':probsp,
            'Matriz de Probabilidades':mostrarMatriz}

    return output
```

```python
for i in range(jogos.shape[0]):
    selecao1 = jogos['seleção1'][i]
    selecao2 = jogos['seleção2'][i]
    v, e, d = ProbabilidadesPartida(selecao1, selecao2)['Probabilidades']
    jogos.at[i, 'Vitória'] = v
    jogos.at[i, 'Empate'] = e
    jogos.at[i, 'Derrota'] = d
```
